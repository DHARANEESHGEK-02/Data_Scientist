Matrix Multiplication

```python
import numpy as np
# 2x2 matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Method 1: np.matmul()
result = np.matmul(A, B)
print(result)
# Output: [[19 22]
#          [43 50]]

# Method 2: @ operator (Python 3.5+)
result = A @ B
print(result)  # Same output

```
