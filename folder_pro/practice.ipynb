Price books problem

```python
def find_pb(price,eps,book):
    pe=price/eps
    pb=price/book
    return pe,pb
find_pb(100,2,4)
```

Problems Using Dictionary
1)Access through for loop

```python
my_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}


my_dict['country'] = 'USA'
my_dict['age'] = 26 

# Method 1
for k in my_dict:
    print(f"{k}: {my_dict[k]}")
    
# Method 2
for k ,j in my_dict.items():
    print(k,j)
```

```python
# Safe access with get()
print(my_dict.get('age', 'Unknown'))  # 25
print(my_dict.get('country', 'Unknown'))  # Unknown
```
