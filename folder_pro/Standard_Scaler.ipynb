```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
import numpy as np

# Sample data with 5 features
X = np.array([[1, 1000, 10, 20, 5], [2, 2000, 15, 25, 3], [3, 3000, 20, 30, 7], [4, 4000, 25, 35, 2]])
y = np.array([0, 1, 0, 1])

# Step 1: Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Step 2: Fit scaler ONLY on TRAIN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 3: Feature selection ONLY on TRAIN
selector = SelectKBest(f_regression, k=2)
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)

print("Selected features:", np.arange(5)[selector.get_support()])
print("Train shape after selection:", X_train_selected.shape)  # (3, 2)
```
