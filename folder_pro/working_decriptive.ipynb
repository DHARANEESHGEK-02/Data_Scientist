```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
np.random.seed(42)
data={
    "Math_mark":np.random.randint(45,100,25),
    "Science_mark":np.random.randint(47,100,25)
    }
df=pd.DataFrame(data)
print("Dataset:\n",df)
```

```python
decs_stats=df.describe()
print("Descriptive Stastics:\n",decs_stats)
```

```python
# Central Tendency

central_tend={
    "Mean:":df.mean(),
    "Median:":df.median(),
    "Mode:":df.mode().iloc[0],
    
}

print("\n Central Tendency:\n",pd.DataFrame(central_tend))
```

```python
# --- Measures of Dispersion ---
dispersion = {
    "Variance": df.var(),
    "Standard Deviation": df.std(),
    "Range": df.max() - df.min(),
    "IQR": df.quantile(0.75) - df.quantile(0.25)
}
print("\nDispersion Measures:\n", pd.DataFrame(dispersion))
```

```python
plt.figure(figsize=(10,5))
sb.histplot(df["Math_mark"],bins=6,kde=True,color='Blue',edgecolor='Black')
plt.title("Histogram Using Math Score")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.grid()
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Box Plot for Science Score
plt.figure(figsize=(6, 4))
sb.boxplot(x=df["Science_mark"], color='lightgreen')
plt.title("Box Plot of Science Scores")
plt.xlabel("Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```
