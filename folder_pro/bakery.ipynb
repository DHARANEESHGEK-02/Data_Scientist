```python
import numpy as np
import pandas as pd

# 1. Create a NumPy array of scores
scores = np.array([92, 85, 76, 88, 95, 67, 73, 81, 59, 90])

# 2. Basic statistics with NumPy
mean_score = np.mean(scores)
max_score = np.max(scores)
min_score = np.min(scores)

print("NumPy stats:")
print("Mean:", mean_score)
print("Max :", max_score)
print("Min :", min_score)

# 3. Convert to pandas DataFrame
df = pd.DataFrame({"Score": scores})

# 4. Add a grade column using pandas
def grade(x):
    if x >= 90:
        return "A"
    elif x >= 80:
        return "B"
    elif x >= 70:
        return "C"
    elif x >= 60:
        return "D"
    else:
        return "F"

df["Grade"] = df["Score"].apply(grade)

# 5. Show summary by grade
print("\nPandas grade summary:")
print(df["Grade"].value_counts())
```
